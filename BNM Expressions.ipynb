```python
import json
import os
import bnm
import re
```

```python
def read_json(file):
    json_file = open(file,encoding='utf-8').read()
    json_data = json.loads(json_file)
    return json_data
```

```python
books = read_json('expressions.json')
print(len(books))
```

```python
data = dict()

out = open('bnm_expressions.tsv','w',encoding='utf-8')
out.write('locus_incipit\tlocus_explicit\tsequence\tpid\tannotation\tlanguage\tsame_line\n')
    

for book in books:
    print(book)

    texts = books[book]
    total_number = len(texts)
    
    path = os.path.join('BNM_carriers',book)
    book_data = read_json(path)

    text_sequence = dict()
    
    if 'contains_text' in book_data['@relations']:
        locus_text = dict()
        bnm_texts = book_data['@relations']['contains_text']
        for t in bnm_texts:
            seq = t.get('postText','')
            if re.search('[-]',str(seq)):
                seq = seq[:seq.index('-')]
            seq = re.sub("\D", "", str(seq))
            locus_text[seq] = f"{t['id']}.json"
            
        for i,text in enumerate(sorted(locus_text)):
            text_sequence[locus_text[text]] = i+1
            print('f.' + text , text_sequence[locus_text[text]],'/',total_number)
    

    for sequence,text in enumerate(sorted(texts)):
        
        data = dict()
        path = os.path.join('BNM_texts',text)
        text_data = read_json(path)
         
            
        if 'postText' in text_data['@relations']['text_part_of'][0]:
            data['locus'] = text_data['@relations']['text_part_of'][0]['postText']

            if re.search( r'[-]' , str(data['locus']) ):
                parts = re.split( r'[-]' , str(data['locus']) )
                data['locus_incipit'] = parts[0]
                data['locus_incipit'] = re.sub( r'f\.+s*' , '' , data['locus_incipit'] )

                data['locus_explicit'] = parts[1]
                data['locus_explicit'] = re.sub( r'f\.+s*' , '' , data['locus_explicit'] )

        sequence = text_sequence.get(text,'?')
        data['sequence'] = f'{sequence}/{total_number}'

        if 'datering' in text_data and text_data['datering'] is not None:
            data['date'] = text_data['datering']
            if re.search(r'[/]',str(data.get('date')) ):
                parts = re.split(r'[/]',data['date'])
                data['date'] = parts[0]

        if text_data['_id'] is not None:
            data['pid'] = text_data['_id']
        languages = bnm.get_languages(text_data)

        if len(languages)>0:
            
            if len(languages[0])>0:
                data['language'] = languages[0]
                print(data['language'])

        notes = bnm.get_annotations(text_data)
        data['annotation'] = ' '.join(notes) 

        
        ms_desc = bnm.get_overige_hs_aanduiding(text_data)


        out.write( f"{data.get('locus_incipit','')}\t{data.get('locus_explicit','')}\t{data['sequence']}\t{data['pid']}\t{data['annotation']}\t{data.get('language','')}\tNo\n" ) 
        
out.close()
print('Done!')
```
